```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
%matplotlib inline
```

```python
df=pd.read_csv(r"C:\Users\Dell\Desktop\data\canada_per_capita_income.csv")
df
```

```python
df.rename(columns={'per capita income (US$)':'percapita'},inplace=True)
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('year')

plt.ylabel('percapita')
plt.scatter(df.year,df.percapita,color="red",marker="*")
```

```python
nyr=df.drop('percapita',axis='columns')
nyr
```

```python
percapita1=df.percapita
percapita1
```

```python
reg=linear_model.LinearRegression()
reg.fit(nyr,percapita1)
```
